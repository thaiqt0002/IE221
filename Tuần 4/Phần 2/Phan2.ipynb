```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

## Bài 1

```python
#a, Lập bảng tần số

df = pd.read_csv('dataset/bai1.csv')

frequency = df['productivity'].value_counts()
print(frequency)
```

```python
#b, Vẽ biểu đồ cột biểu diễn sự phân bố tần suất
frequency.plot(kind='bar', width=0.8)
plt.xlabel('Productivity')
plt.xticks(rotation=0)
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Productivity')
plt.show()
```

```python
#c,  Tính số trung bình, độ lệch chuẩn, số trung vị và mốt
average_productivity = df['productivity'].mean()
std_dev_productivity = df['productivity'].std()
med_productivity = df['productivity'].median()
mode_productivity = df['productivity'].mode()
print("Average Productivity: ", average_productivity)
print("Standard Deviation of Productivity: ", std_dev_productivity)
print("Median Productivity: ", med_productivity)
print("Mode of Productivity: ", mode_productivity)
```

## Bài 5

```python
#Nhập dữ liệu 
df = pd.read_csv('dataset/bai5.csv')
df.head()
```

```python
average_before_2011 = df.query('2001 <= year <= 2010')['score'].mean()
average_after_2011 = df.query('2011 <= year <= 2020')['score'].mean()

median_before_2011 = df.query('2001 <= year <= 2010')['score'].median()
median_after_2011 = df.query('2011 <= year <= 2020')['score'].median()

print('Trung bình điểm của 2 giai đoạn: ', average_before_2011, average_after_2011)
print('Trung vị điểm của 2 giai đoạn:', median_before_2011, median_after_2011)
```

Nhận xét: Điểm thi giai đoạn trước đều có cả trung bình điểm lẫn trung vị điểm lớn hơn giai đoạn sau. Nên có thể cho rằng điểm thi của đội tuyển giai đoạn 2001 - 2010 cao hơn giai đoạn 2011 - 2020

## Bài 20

```python
df = pd.read_csv('dataset/bai20.csv') 
```

```python
correlation = df['temperature'].corr(df['weight'])
print('Hệ số tương quan:', correlation)
```

```python
# Assume df is your DataFrame and it includes columns 'x' and 'y'
X = df[['temperature']]
y = df['weight']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the coefficient (a) and intercept (b)
a = model.coef_[0]
b = model.intercept_

print(f"The regression line is: y = {a} * x + {b}")

# Plotting
plt.scatter(df['temperature'], df['weight'], color='b')
plt.plot(df['temperature'], a * df['temperature'] + b, color='r')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# Assume `model` is your trained linear regression model
temperature_df = pd.DataFrame([1.75], columns=['temperature'])
sugar_mass = model.predict(temperature_df)
print(f"The estimated sugar mass for a temperature of {temperature_df.values[0][0]}C is {sugar_mass[0]}")
```
